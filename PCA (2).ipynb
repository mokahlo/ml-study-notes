### Import Libraries

```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
```

### Read data using Pandas

```python
train_path = 'breast_cancer_train.csv'
```

```python
df = pd.read_csv(train_path)
print(df.shape)
df.head()
```

### Note: The target columns in not the last column here.

### Drop unnecessary columns (id)

```python
df.drop(columns='id', inplace=True)
```

### Binarize the target column: Change from B/M -> 0/1.

```python
lb = preprocessing.LabelBinarizer()
df[df.columns[0]] = lb.fit_transform(df[df.columns[0]])
df.head()
```

### Split the data into X and y

```python
X = df[df.columns[1:]]
y = df[df.columns[0]]
```

### Scale the data using Standard Scaler

```python
scaler = StandardScaler()

X[X.columns] = scaler.fit_transform(X)
```

```python
X.shape
```

### Use PCA to reduce number of dimensions
https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html

#### Fit the training data and store in X_pca. Use whiten = True, 3 components, and print the shape

```python
from sklearn.decomposition import PCA


PCAinst=PCA(n_components=3, whiten=True)
X_PCA= PCAinst.fit_transform(X)
```
